```python
from gridProduction import getData
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline  
data=getData()
print(data.info())
```

```python
stateCoulums=[x for x in data.columns if ('total' not in x and 'United' not in x and 'Columbia' not in x)]
print(len(stateCoulums))
data[stateCoulums].plot(figsize=(15,20))


```

```python
import seaborn as sns

fig, ax = plt.subplots(figsize=(15,15))
ax = sns.heatmap(data[stateCoulums].corr())
```

```python
import numpy as np
from sklearn import decomposition
from sklearn import datasets

pca = decomposition.PCA(n_components=3)
X=np.array(data[stateCoulums]).T
X=X/np.mean(X,axis=1).reshape((50,1))
pca.fit(X)
Y= pca.transform(X)

fig, ax = plt.subplots(figsize=(15,15))
ax.scatter(Y[:,0],Y[:,1])
for i in range(len(stateCoulums)):
    ax.annotate(stateCoulums[i], (Y[i,0],Y[i,1]))

```
